```
import pandas as pd
import numpy as np
import random

# create a new dataframe
msft_wma = pd.DataFrame({'colA': random.choices(range(1, 7), k=10)})

# calculate the weighted moving average and add it to the dataframe
weights = np.array([0.5, 0.25, 0.25])
sum_weights = np.sum(weights)

msft_wma['weighted_ma'] = (msft_wma['colA']
    .rolling(window=3, center=True)
    .apply(lambda x: np.sum(weights*x) / sum_weights, raw=False)
)

# print the dataframe before concatenation
print("Before concatenation:")
print(msft_wma)

# create another dataframe to concatenate
new_data = pd.DataFrame({'colA': random.choices(range(1, 7), k=5)})

# calculate the weighted moving average for the new data
new_data['weighted_ma'] = (new_data['colA']
    .rolling(window=3, center=True)
    .apply(lambda x: np.sum(weights*x) / sum_weights, raw=False)
)

# concatenate the new dataframe with the existing one
msft_wma = pd.concat([msft_wma, new_data])

# print the concatenated dataframe
print("After concatenation:")
print(msft_wma)
```
