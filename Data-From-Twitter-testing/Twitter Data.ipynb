```
import pandas as pd
import datetime
df = pd.read_csv('/datasets/achintyas-google-drive/Dataset/Tweet.csv', engine='python')

df.head()
```

```
df.shape
```

```
import re
counter = 0
contains_msft = []
for tweet in df['body']:
    if re.search('microsoft|MSFT', tweet):
        counter+=1
        contains_msft.append('y')
    else:
        contains_msft.append('n')
print(counter)
df['contains_microsoft'] = contains_msft
df.head()
#adding a column to the df that has a y or n if the tweet contains microsoft or not. so we can remove
#the tweets that dont
```

```

df=df.drop(df[df['contains_microsoft'] == 'n'].index)
df.head()

#df is now only entries about microsoft
```

```
#interaction score- the interaction score formula includes the authors number of tweets and followers
#I think it might be interaction score for an account?
#I'm going to sum up comments + likes + quotes and minmax scale that.
```

```
df['interaction_num'] = df[['comment_num','retweet_num','like_num']].sum(axis=1)
df=df.drop(columns=['comment_num','like_num','retweet_num','contains_microsoft'])
df.head()
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['interaction_num']])
df['interaction_num']=scaler.transform(df[['interaction_num']])
df.head()
```

```
df['post_date'] = pd.to_numeric(df['post_date'], errors='coerce')
df = df.dropna()

df.head()
```

```
from datetime import datetime, timedelta
date_ints = df['post_date']
#start_date = datetime(2000, 1, 1)

dates = []

for date in date_ints:
    dt = datetime.fromtimestamp(date)
    dates.append(dt)

```

```
print(date_ints[:5])
print(dates[:5])
```

```
df['date'] = dates
df.head()
```

```
df['tweet_date'] = df['date'].dt.date

df.head()
```

```
df = df.drop(columns=['post_date','date'])
```

## Sentiment Analysis

```
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text

    tokens = word_tokenize(text.lower())




    # Remove stop words

    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]




    # Lemmatize the tokens

    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]




    # Join the tokens back into a string

    processed_text = ' '.join(lemmatized_tokens)

    return processed_text

# apply the function df

df['body'] = df['body'].apply(preprocess_text)
df.head()
```

```


# initialize NLTK sentiment analyzer

analyzer = SentimentIntensityAnalyzer()


# create get_sentiment function

def get_sentiment(text):

    scores = analyzer.polarity_scores(text)

    sentiment = 1 if scores['pos'] > 0 else 0

    return sentiment




# apply get_sentiment function

df['sentiment'] = df['body'].apply(get_sentiment)

df.head()
```

```
df.to_pickle('twitter_sentiment.pkl')

sentiment = df['sentiment'].to_list()

file = open('setiments.txt','w')
for item in sentiment:
	file.write(str(item) +"\n")
file.close()
```

```
grouped = df.groupby('tweet_date').agg({'sentiment': 'mean', 'interaction_num': 'mean'})

grouped.head()
```
