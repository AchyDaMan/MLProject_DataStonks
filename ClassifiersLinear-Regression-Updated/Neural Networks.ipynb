```
import pandas as pd
import numpy as np

msft_official_features = pd.read_pickle('final_features.pkl')
msft_official_labels = pd.read_pickle('outcome.pkl')
```

```
msft_official_features.columns
```

```
msft_official_labels = pd.DataFrame(msft_official_labels)
```

```
train_features = msft_official_features[['past_open', 'past_high', 'past_low', 'past_close', 'past_volume',
       'past_stock_splits', 'curr_open', 'SMA_10', 'SMA_50', 'SMA_100',
       'sentiment', 'interaction_num']]
train_labels = msft_official_labels[['outcome']]
train_labels = train_labels.values.tolist()

train_features = train_features.to_numpy()
train_labels = np.array(train_labels)

type(train_labels)
#print(train_labels)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import set_random_seed

set_random_seed(555)
total_num_classes = 1

model = Sequential()
model.add(Dense(4, input_shape=(12,), activation='linear'))
model.add(Dense(8, activation='tanh'))
model.add(Dense(4, activation='tanh'))
model.add(Dense(total_num_classes, activation='softmax'))

model.summary()
```

```
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam

optim = Adam(learning_rate=0.05) 
loss_fn = CategoricalCrossentropy()

model.compile(loss=loss_fn,optimizer=optim,metrics=["accuracy"])
```

```
model.fit(train_features, train_labels, epochs=100, batch_size=8)
```
